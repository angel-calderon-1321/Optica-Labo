# Convolucion de señales

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

Definimos las funcoines a convolucionar

```python
def escalon(x):
    return  np.heaviside(x,1)

def rect(x, width = 0.5):
    return  np.where(np.abs(x) <= width, 1, 0)  
```

```python
x = np.linspace(-10, 10, 1000)
# Las dos funciones a convolucionar
h_vals = rect(x)
g_vals = rect(x)
a = -0.6
g_vals_tras = rect(a-x)
```

```python
#Resultado de la convolución
conv_result = signal.convolve(h_vals, g_vals, mode='same')/sum(g_vals)
```

```python
# Inicializamos la figura
fig, axs = plt.subplots(2)
axs[0].plot(x, h_vals, 'r', lw=2, label=r'step($\xi$)')
axs[0].plot(x, g_vals_tras, 'b', lw=2, label=r'rect(x-$\xi$)')

axs[1].plot(x, conv_result, 'k', lw=2, label='h * g (convolución)')
 
axs[0].set_title("Señal Originial y Filtro")
axs[0].legend()
axs[0].set_ylim([0, 1.5])
axs[0].set_xlim([-3, 3])
axs[0].set_xlabel(r'$\xi$')
axs[0].grid()

axs[1].set_title("Señal Filtrada")
axs[1].set_ylim([0, 1.5])
axs[1].set_xlim([-3, 3])
axs[1].set_xlabel('x')
axs[1].legend()
axs[1].grid()

fig.tight_layout()
```

# Convolve  Animation

```python
#animacion de la convolucion

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
# Definimos la función
def plot_sine(shift=0):
    x = np.linspace(-10, 10, 400)
    y = np.sin(x - shift)   # desplazamiento horizontal
    plt.figure(figsize=(8,4))
    plt.plot(x, y, label=f'sin(x - {shift:.2f})')
    plt.axhline(0, color='k', linewidth=0.8)
    plt.axvline(0, color='k', linewidth=0.8)
    plt.legend()
    plt.grid(True)
    plt.ylim(-1.2, 1.2)
    plt.show()

# Deslizador interactivo
interact(plot_sine, shift=(-5, 5, 0.1))
# Mueve el deslizador para ver cómo cambia la función seno con el desplazamiento horizontal.
# Puedes ajustar los límites y el paso del deslizador según tus necesidades.

#plt.close()  # Cierra la figura para evitar que se muestre estática al final
```
